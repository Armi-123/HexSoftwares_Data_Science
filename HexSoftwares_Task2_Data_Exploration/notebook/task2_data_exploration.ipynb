# Task 2: Data Exploration using Iris Dataset

## Objective
Explore the Iris dataset, check data quality, and visualize feature distributions.

## Tools Used
- Python
- Pandas
- Matplotlib
- Seaborn
- Jupyter Notebook

```python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
# Load dataset

df = pd.read_csv("../data/Iris.csv")

df.head()
```

## Dataset Overview

```python
df.info()

df.describe()
```

## Missing Values Analysis

```python
df.isnull().sum()
```

## Distribution of Features

```python
df.hist(figsize=(10,8))
plt.tight_layout()
plt.savefig("../images/histogram.png")
plt.show()
```

## Boxplots for Outlier Detection

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df.iloc[:,1:5])
plt.savefig("../images/boxplot.png")
plt.show()
```

## Pairwise Feature Relationships

```python
sns.pairplot(df, hue="Species")
plt.savefig("../images/pairplot.png")
plt.show()
```

## Correlation Analysis

```python
plt.figure(figsize=(8,6))
sns.heatmap(
    df.iloc[:,1:5].corr(),
    annot=True,
    cmap="coolwarm"
)
plt.savefig("../images/heatmap.png")
plt.show()
```

## Key Insights

- Petal Length and Petal Width show strong correlation.
- Setosa species is clearly separated from others.
- No missing values found in the dataset.
- Feature distributions are nearly normal.
- Some features contain minor outliers.

## Conclusion

The Iris dataset was successfully explored and visualized. The analysis revealed strong relationships among petal features and clear separation among species classes.
